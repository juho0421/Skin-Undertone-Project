# 2. mask 함수 설정

```python
# 색보정 미적용 mask 
def mask(image, name):
  face_cascade = cv2.CascadeClassifier(path_for_face)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # YCrCb 색 공간으로 변환
  ycbcr_image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

    # 필터링할 색상 범위 정의
  lower_bound = np.array([0, 133, 77], dtype=np.uint8)
  upper_bound = np.array([255, 173, 127], dtype=np.uint8)

    # 색상 범위에 해당하는 마스크 생성
  mask = cv2.inRange(ycbcr_image, lower_bound, upper_bound)

    # 마스크를 이용하여 이미지에서 해당 색상만 남기고 나머지 부분은 검은색으로 처리
  filtered_image = cv2.bitwise_and(image, image, mask=mask)

  for (x, y, w, h) in faces:
    # 이미지 자르기
    cropped = filtered_image[y: y+h, x: x+w]
    resize = cv2.resize(cropped, (180,180))

    # 이미지 저장경로
    cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/머신러닝1/웜쿨 프로젝트/cool tone_adjusted/'+name+'.jpg', resize) # 이미지 파일 저장 경로로 바꾸기

# face에 저장된 사진을 mask에 저장
for name in file_name_list:
  image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/머신러닝1/웜쿨 프로젝트/cool tone/'+name+'.jpg') # 이미지 저장 경로로 바꾸기
  mask(image, name)
```

3.1. 히스토그램 균일화 보정

```python
def adjust_brightness(image, name):
  face_cascade = cv2.CascadeClassifier(path_for_face)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  faces = face_cascade.detectMultiScale(gray, 1.3, 5)
  for (x, y, w, h) in faces:

    # YCrCb 색 공간으로 변환
    img_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
    
    # Y 채널만 히스토그램 균일화 적용
    img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
    
    # 다시 BGR 색 공간으로 변환
    img_bgr = cv2.cvtColor(img_yuv, cv2.COLOR_YCrCb2BGR)
    
    # 이미지 저장경로
    cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/머신러닝1/웜쿨 프로젝트/cool tone_brightness/'+name+'.jpg', img_bgr) # 이미지 파일 저장 경로로 바꾸기

# mask에 저장된 사진을 mask2에 저장
for name in file_name_list2:
  image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/머신러닝1/웜쿨 프로젝트/cool tone_adjusted/'+name+'.jpg') # 이미지 파일 경로로 바꾸기
  adjust_brightness(image, name)
```

3.2. 화이트 밸런스

```python
def white_balance(image, name):
    result = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    avg_a = np.average(result[:, :, 1])
    avg_b = np.average(result[:, :, 2])
    for x in range(image.shape[0]):
        for y in range(image.shape[1]):
            l, a, b = result[x, y, :]
            result[x, y, 1] = a - ((avg_a - 128) * (l / 255.0) * 1.1)
            result[x, y, 2] = b - ((avg_b - 128) * (l / 255.0) * 1.1)
    result = cv2.cvtColor(result, cv2.COLOR_LAB2BGR)

    cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/머신러닝1/웜쿨 프로젝트/cool tone_whitebalanced/'+name+'.jpg', result) # 이미지 파일 저장 경로로 바꾸기


for name in file_name_list2:
  image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/머신러닝1/웜쿨 프로젝트/cool tone_adjusted/'+name+'.jpg') # 이미지 파일 경로로 바꾸기
  white_balance(image, name)
```

# 4. 데이터 증강

```python
import os
import random
import cv2
from PIL import Image
import numpy as np

def augment_images(image_dir, save_dir):
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    file_names = os.listdir(image_dir)

    for filename in file_names:
        image_path = os.path.join(image_dir, filename)
        image = Image.open(image_path)
        
        raw_image = image
        raw_filename = f"raw_{filename}"
        raw_save_path = os.path.join(save_dir, raw_filename)
        raw_image.save(raw_save_path)

        # 좌우 반전 이미지 생성
        flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)
        flipped_filename = f"flipped_{filename}"
        flipped_save_path = os.path.join(save_dir, flipped_filename)
        flipped_image.save(flipped_save_path)

        # 기울이기 이미지 생성
        angle = random.randint(-30, 30)  # 기울이기 각도 설정
        rotated_image = image.rotate(angle)
        rotated_filename = f"rotated_{filename}"
        rotated_save_path = os.path.join(save_dir, rotated_filename)
        rotated_image.save(rotated_save_path)

        # 블러 처리 이미지 생성
        image_array = np.array(image)
        image_array = image_array.astype(np.uint8)  # 데이터 타입 변환
        ksize = (5, 5)  # 필터 크기
        sigma = 10  # 시그마 값
        blurred_image = cv2.GaussianBlur(image_array, ksize, sigma)
        blurred_image_pil = Image.fromarray(blurred_image)
        blurred_filename = f"blurred_{filename}"
        blurred_save_path = os.path.join(save_dir, blurred_filename)
        blurred_image_pil.save(blurred_save_path)

# 사용 (경로 변경)
image_dir = '/content/drive/MyDrive/Colab Notebooks/머신러닝1/웜쿨 프로젝트/warm tone_whitebalanced'
save_dir = '/content/drive/MyDrive/Colab Notebooks/머신러닝1/웜쿨 프로젝트/warm tone_argumentation'

augment_images(image_dir, save_dir)
```
